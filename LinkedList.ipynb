# Linked List Notes

## What is a Linked List?

A linked list is a data structure consisting of nodes. Each node contains two elements:

1. **Value:** The data stored in the node.
2. **Next:** A reference (or pointer) to the next node in the sequence.

## How to Connect Nodes in a Linked List

To create a linked list, you connect multiple nodes together:

```python
# Example of a simple linked list

# Create nodes
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

node1 = Node('red')
node2 = Node('blue')
node3 = Node('green')

# Link nodes together
node1.next = node2
node2.next = node3
node3.next = None  # Last node points to None


Head: The first node of the linked list is referred to as the head. In this example, node1 is the head of the linked list.
Characteristics of Linked Lists
Non-contiguous Memory Allocation:

Linked lists are not stored in contiguous memory locations like arrays. Instead, nodes are scattered throughout memory, and they are connected through pointers.
Dynamic Size:

Linked lists can easily grow or shrink in size by adding or removing nodes, without the need to allocate or deallocate memory blocks of fixed size.
Comparison with Arrays
Linked Lists:

Use pointers to connect nodes.
Provide flexibility in memory usage.
Allow efficient insertions and deletions.
Arrays:

Stored in contiguous memory locations.
Have fixed size once allocated.
Provide quick access to elements by index.
Summary
Linked lists are useful for dynamic data structures where elements frequently change. Understanding how to connect nodes with pointers is essential for working with linked lists.

````python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

    def get_next(self):
        return self.next

    def set_next(self, n):
        self.next = n

    def get_data(self):
        return self.value

    def set_data(self, d):
        self.value = d


class LinkedList:
    def __init__(self, head=None):
        self.head = head
        self.size = 0

    def get_size(self):
        return self.size

    def add(self, value):
        new_node = Node(value, self.head)
        self.head = new_node
        self.size += 1

    def remove(self, d):
        curr = self.head
        prev = None
        while curr:
            if curr.value == d:
                if prev:
                    prev.next = curr.next
                else:
                    self.head = curr.next
                self.size -= 1
                return True
            else:
                prev = curr
                curr = curr.next
        return False

    def find(self, d):
        curr = self.head
        while curr:
            if curr.value == d:
                return True
            curr = curr.next
        return False

    def print_list(self):
        curr = self.head
        values = []
        while curr:
            values.append(curr.value)
            curr = curr.next
        print("LinkedList:", values)


# Example usage
furqiList = LinkedList()

# Adding elements
furqiList.add(3)
furqiList.add(8)
furqiList.add(6)
furqiList.add(7)

# Print list after additions
print("After additions:")
furqiList.print_list()

# Removing elements
furqiList.remove(8)
furqiList.remove(6)

# Print list after removals
print("After removals:")
furqiList.print_list()
````

##### Reverse a LinkedList

````python
# Definition for singly-linked list node.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # If the list is empty or has only one node, return the head
        if not head or not head.next:
            return head
        
        # Initialize the previous, current, and next pointers
        prev, curr = None, head
        
        while curr:
            # Store the next node
            next_node = curr.next
            # Reverse the current node's pointer
            curr.next = prev
            # Move pointers one step forward
            prev = curr
            curr = next_node
        
        # Prev will be the new head of the reversed list
        return prev
````
