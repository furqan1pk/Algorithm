```python


from collections import deque

def BFS(AL, src):
    q = deque()
    visited = set()
    q.append((src, ""))  # Initialize queue with source and an empty path
    
    while q:
        current_src, psf = q.popleft()  # Use q, not queue
        
        if current_src not in visited:
            visited.add(current_src)  # Mark as visited
            print(f"Visited {current_src} with path {psf}{current_src}")  # Show the node and path
            
            for nbr in AL[current_src]:
                if nbr not in visited:
                    q.append((nbr, psf + str(current_src)))  # Append current_src to path, not 'src'

# Example adjacency list for a graph
AL = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Run BFS on the graph from source 'A'
BFS(AL, 'A')
```

```python
import heapq
def BFS(AL,src,dest):

    heap=[]

    #heapq.heappush(heap,[src,"",0])
    #heapq.heappush(heap, [weight, src, data])
    heapq.heappush(heap, [0, src, ""])
    visited=set()

    while heap:

        # remove
        w,c_src,psf=heapq.heappop(heap)

        print(f"psf is {psf} and weight is {w}")

        # Mark
        if c_src not in visited:
            visited.add(c_src)
        for neighbor, weight in AL[c_src]:
            if neighbor not in visited:
                heapq.heappush(heap,[w+weight,neighbor,psf+str(neighbor)])



                                     
# Graph represented as an adjacency list
# The keys are the nodes, and the values are lists of tuples (neighbor, weight)
graph = {
    'A': [('B', 4), ('C', 2)],
    'B': [('C', 5), ('D', 10)],
    'C': [('E', 3)],
    'D': [('F', 11)],
    'E': [('D', 4)],
    'F': []
}

# This graph looks like this:
# A --4-- B --10-- D --11-- F
#  \      \       /
#   2      5     4
#    \      \   /
#     C --3-- Es
        

BFS(graph,'A','F')
```

```python
class Solution:
    def nearestExit(self, maze: List[List[str]], entrance: List[int]) -> int:

        def check_bounds(i, j):
            return 0 <= i < len(maze) and 0 <= j < len(maze[0])

        def is_entrance_valid(i,j):
            return maze[i][j]=='.'
            
        def check_exit(i, j):
            return i == 0 or i == len(maze) - 1 or j == 0 or j == len(maze[0]) - 1

        def BFS(entrance):

            visited = set()
            q = deque()
            q.append((entrance[0], entrance[1], 0))
            while q:
                # remove
                i, j, level = q.popleft()
                if (i, j) not in visited:
                    # mark
                    visited.add((i, j))
                    # work
                    if i != entrance[0] or j != entrance[1]:
                        if check_exit(i, j):
                            return level
                    # this is nice way to iterate
                    # Add
                    for x, y in [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]:
                        if check_bounds(x, y) and is_entrance_valid(x, y) and (x, y) not in visited:
                            q.append((x, y, level + 1))
                    
            return -1

        return BFS(entrance)

        #     while q:
        #         # remove
        #         i,j,level=q.popleft()

        #         if (i,j) not in visited:
        #             visited.add((i,j))
        #             # print or work
        #             if i != entrance[0] or j != entrance[1]:
        #                 if check_exit(i, j):
        #                     return level
        #         # check up and add

        #         if is_entrance_valid(i-1,j) and check_bounds(i-1,j):
        #             q.append(i-1,j,level+1)
        #         # check down and add
        #         if is_entrance_valid(i+1,j) and check_bounds(i+1,j):
        #             q.append(i+1,j,level+1)
        #         # check right
        #         if is_entrance_valid(i,j+1) and check_bounds(i,j+1):
        #             q.append(i,j+1,level+1)
        #         #check left
        #         if is_entrance_valid(i,j-1) and check_bounds(i,j-1):
        #             q.append(i,j-1,level+1)
        # BFS(entrance)
                
```
