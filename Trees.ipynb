```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

    def add_child(self, child_node):
        """Adds a child to the current node."""
        self.children.append(child_node)

    def print_tree(self, level=0):
        """Prints the tree structure starting from the current node."""
        indent = ' ' * (level * 2)
        print(indent + str(self.data))
        for child in self.children:
            child.print_tree(level + 1)

    def print_height(self):
        ht=-1
        for child in self.children:
            height =child.print_height()
            ht=max(height,ht)
        return ht+1

    def print_maximum(self):
        maximum=0
        for child in self.children:
            m =child.print_maximum()
            maximum=max(m,maximum)
        maximum=max(self.data,maximum)
        return maximum

    def print_size(self):
        total_size=0
        for child in self.children:
            size =child.print_size()
            total_size=total_size+size
        return total_size+1


    def traversel(self):
        # pre order   before going deep in recursions  (node ki left side pe hota hai)  Node > Child   root is first
        print("node pre"+str(self.data))
        for child in self.children:
            # edgee pre
            print("edge pre"+str(self.data)+"--"+str(child.data))
            child.traversel()
            print("edge post"+str(self.data)+"--"+str(child.data))

            # edge post

        # post order  node ki right side ko kehte haen... while coming out of the recursions   Child > node root is last
        print("node post"+str(self.data))

    def level_order_traversal(self):
        # you can use queue with behavior of remove print and add... while removing hte node u haave to add its children to the queue so they will be printed lvel by level
        """Traverses the tree in level order."""
        if not self:
            return
        
        queue = [self]  # Start with the root node in the queue
        
        while queue:
            current_node = queue.pop(0)  # Dequeue the first element
            print(current_node.data, end=" ")  # Process the current node
            
            # Enqueue child nodes of the current node
            for child in current_node.children:
                queue.append(child)
        
# Example usage
root = TreeNode("Root")
child1 = TreeNode("Child1")
child2 = TreeNode("Child2")

root.add_child(child1)
root.add_child(child2)

child1.add_child(TreeNode("Child1.1"))
child1.add_child(TreeNode("Child1.2"))
child2.add_child(TreeNode("Child2.1"))

root.print_tree()
root.print_height()
root.print_size()
#root.print_maximum()
root.traversel()
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = TreeNode(data)
        else:
            self._insert_recursive(self.root, data)

    def _insert_recursive(self, current, data):
        if data < current.data:
            if current.left is None:
                current.left = TreeNode(data)
            else:
                self._insert_recursive(current.left, data)
        else:
            if current.right is None:
                current.right = TreeNode(data)
            else:
                self._insert_recursive(current.right, data)

    def inorder_traversal(self):
        result = []
        self._inorder_recursive(self.root, result)
        return result

    def _inorder_recursive(self, current, result):
        if current:
            self._inorder_recursive(current.left, result)
            result.append(current.data)
            self._inorder_recursive(current.right, result)

# Example usage:
my_tree = BinaryTree()
my_tree.insert(50)
my_tree.insert(30)
my_tree.insert(20)
my_tree.insert(40)
my_tree.insert(70)
my_tree.insert(60)
my_tree.insert(80)

inorder_result = my_tree.inorder_traversal()
print(inorder_result)
```
