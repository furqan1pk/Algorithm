```python
from typing import Optional

class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None

    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        previous = None
        current = head

        while current:
            save_next = current.next
            current.next = previous
            previous = current
            current = save_next

        return previous

    def display(self):
        current_node = self.head
        while current_node:
            print(current_node.value, end=" -> ")
            current_node = current_node.next
        print("None")

# Example usage:
linked_list = LinkedList()
linked_list.head = ListNode(10, ListNode(20, ListNode(30, ListNode(40))))

print("Original Linked List:")
linked_list.display()

linked_list.head = linked_list.reverseList(linked_list.head)

print("\nReversed Linked List:")
linked_list.display()
```
