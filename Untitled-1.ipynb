```python
def solveNQueens(n):
    chess = [[0 for _ in range(n)] for _ in range(n)]
    solutions = []

    def issafetoputqueen(rows, column):
        # check upward direction if any queen already present
        i = rows - 1
        j = column
        while i >= 0:
            if chess[i][j] == 1:
                return False
            i -= 1

        # check diagonal left
        i = rows - 1
        j = column - 1
        while i >= 0 and j >= 0:
            if chess[i][j] == 1:
                return False
            i -= 1
            j -= 1

        # check diagonal right
        i = rows - 1
        j = column + 1
        while i >= 0 and j < n:
            if chess[i][j] == 1:
                return False
            i -= 1
            j += 1
        return True

    def printNqueens(row):
        if row == n:
            # Process and store the current board state
            formatted_solution = []
            for i in range(n):
                row_solution = "".join(["Q" if chess[i][j] == 1 else "." for j in range(n)])
                formatted_solution.append(row_solution)
            solutions.append(formatted_solution)
            return
        for col in range(n):
            if issafetoputqueen(row, col):
                chess[row][col] = 1
                printNqueens(row + 1)
                chess[row][col] = 0

    printNqueens(0)
    return solutions

# Example usage
n = 3
solutions = solveNQueens(n)
print(f"There are {len(solutions)} solutions for {n}-Queens problem.")
for solution in solutions:
    for row in solution:
        print(row)
    print("")  # Print an empty line between solutions
```

```python
def canKnightsAttackEachOther(knight_positions, m, n):
    knight_moves = [(-2, -1), (-2, 1), (-1, -2), (-1, 2), (1, -2), (1, 2), (2, -1), (2, 1)]
    for i, (row1, col1) in enumerate(knight_positions):
        for row2, col2 in knight_positions[i+1:]:
            if (row1, col1) != (row2, col2) and (abs(row1 - row2), abs(col1 - col2)) in knight_moves:
                return True
    return False

def countKnightPlacements(m, n):
    count = 0
    positions = [(row, col) for row in range(m) for col in range(n)]
    for i in range(len(positions)):
        for j in range(i+1, len(positions)):
            for k in range(j+1, len(positions)):
                if not canKnightsAttackEachOther([positions[i], positions[j], positions[k]], m, n):
                    count += 1
    return count

# Example usage
m = 3
n = 3
print(f"Number of ways to place 3 knights on a {m}x{n} board without them attacking each other: {countKnightPlacements(m, n)}")
```

```python
def solveKnights(m, n, knights=3):
    board = [[0 for _ in range(n)] for _ in range(m)]
    solutions = []

    # Check if a knight can be placed at (row, col) without attacking other knights
    def isSafe(row, col):
        directions = [(-2, -1), (-2, 1), (-1, -2), (-1, 2), (1, -2), (1, 2), (2, -1), (2, 1)]
        for dr, dc in directions:
            r, c = row + dr, col + dc
            if 0 <= r < m and 0 <= c < n and board[r][c] == 1:
                return False
        return True

    # Convert board to readable format
    def formatSolution():
        return [['Q' if cell == 1 else '.' for cell in row] for row in board]

    # Backtracking function
    def placeKnights(k, startRow, startCol, count):
        if k == 0:
            solutions.append(formatSolution())
            return True
        else:
            placed = False
            for i in range(startRow, m):
                for j in range(startCol if i == startRow else 0, n):
                    if isSafe(i, j):
                        board[i][j] = 1
                        if placeKnights(k - 1, i, j + 1, count + 1):
                            placed = True
                        board[i][j] = 0
                startCol = 0  # Reset startCol for new rows
            return placed

    placeKnights(knights, 0, 0, 0)
    return len(solutions)

# Example usage
m = 4
n = 4
print(f"Number of ways to place 3 knights on a {m}x{n} board without them attacking each other: {solveKnights(m, n)}")
```
